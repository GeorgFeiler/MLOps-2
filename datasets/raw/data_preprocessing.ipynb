```python
import pandas as pd
```

```python
train = pd.read_csv("train.csv", index_col=0)
test = pd.read_csv("test.csv", index_col=0)
```

```python
train['Sex'] = train['Sex'].apply(lambda x: 1 if x == 'male' else 0)
test['Sex'] = test['Sex'].apply(lambda x: 1 if x == 'male' else 0)

train['Age'] = train['Age'].fillna(train['Age'].mean())
test['Age'] = test['Age'].fillna(train['Age'].mean())
```

```python
features_list = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch']
```

```python
train[features_list].to_csv("~/git/MLOps-2/data/baselines/train.csv", columns=features_list)
```

```python
test[features_list].to_csv("~/git/MLOps-2/data/baselines/test.csv", columns=features_list)
```
